```python
from selenium import webdriver
from bs4 import BeautifulSoup 
from time import sleep
```

```python
driver = webdriver.Chrome("E:\\Lerning\\ML_project\\tools\\chromedriver")
```

```python
driver.get("https://www.olx.in/motorcycles_c81")
```

```python
li = driver.find_elements_by_class_name("EIR5N")
len(li)
```

```python
brand = []
model = []
year = []
driven = []
price = []
discription = []
short_dis = []

for i in range(1,1000):
    back = True
    try:
        li = driver.find_elements_by_class_name("EIR5N")
        li[i].click()
    except:
        driver.get("https://www.olx.in/motorcycles_c81")
        sleep(3)
        for l in range(22,500,20):
            sleep(2)
            driver.find_element_by_xpath("//*[@id='container']/main/div/div/section/div/div/div[5]/div[2]/div/div[2]/ul/li["+str(l)+"]/div/button").click()
            sleep(3)

        li = driver.find_elements_by_class_name("EIR5N")
        li[i].click()
        print(li[i])
    

    try:
        sleep(3)
        el = driver.find_element_by_class_name("CBG3S")
        soup = BeautifulSoup(el.get_attribute("innerHTML"))
    except Exception as e:
        print(e,"  currunt Url:  ",driver.current_url)
        
        back = False
        driver.back()
        

    
    if back:
        try:
            pEl = driver.find_element_by_class_name("_2wMiF")
            pSoup = BeautifulSoup(pEl.get_attribute("innerHTML"))
            p = pSoup.find('span').text.replace('\n','')
            price.append(p)
        except:
            price.append(20000)
        try:
            sd = pSoup.find('h1').text.replace('\n','')
            short_dis.append(sd)
        except:
            short_dis.append("None")

        try:
            b = soup.find('span',attrs={"data-aut-id":"value_make"}).text.replace('\n','')
            brand.append(b)
        except:
            brand.append('None')
        try:
            m = soup.find('span',attrs={"data-aut-id":"value_model"}).text.replace('\n','')
            model.append(m)
        except:
            model.append('None')
        try:
            drive = soup.find('span',attrs={"data-aut-id":"value_mileage"}).text.replace('\n','')
            driven.append(drive)
        except:
            driven.append('0')

        try:
            y = soup.find('span',attrs={"data-aut-id":"value_year"}).text.replace('\n','')
            year.append(y)
        except:
            year.append('None')

        try:
            d = soup.find('div',attrs={"data-aut-id":"itemDescriptionContent"}).text.replace('\n')
            discription.append(d)
        except:
            discription.append('None')
        driver.back()
        print("brand: ",b,"Model: ",m,"price: ",p,"I: ",i)
        
        sleep(3)
    
```

```python
import pandas as pd
```

```python
data = pd.DataFrame({"brand":brand,"model":model,"KM":driven,"year":year,"price":price,"deiscription":discription,"sd":short_dis})
data
```

```python
data1 = pd.read_csv("olx.csv")
```

```python
data=pd.concat([data1,data],ignore_index=True)
data.keys()
```
